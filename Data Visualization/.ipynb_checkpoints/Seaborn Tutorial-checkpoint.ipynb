## 1 - Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# distplot() : shows the distribution of a univariate set of observations. takes a single column of the dataset
# the line on the histogram is call a kde(kernel density estimator)
#sns.distplot(tips['total_bill'])

# remove the kde
#sns.distplot(tips['total_bill'],kde = False)

# change the bins
sns.distplot(tips['total_bill'],kde = False, bins = 30)
```

```python
# joinplot(): allows to match up two distributions plots for bivariate data, combines two distribution plots.
# kind = 'reg','kde', hex','scatter:default' parameter tells us how do we want to compare the two distributions
# x = the first var to be compared, y = the second var, data = data

#sns.jointplot(x = 'total_bill', y = 'tip', data = tips)

# sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')

# sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex') good, it highlights where the points match up the most.
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde') # it shows the density of where the pts match up the most.
```

```python
# Pairplot(): pairwise relationship across the entire dataset. hue arg for relationship between categorical variables and the num var.
# palette arg gives a particular color to the categories. var arg (var = [nameofvariables,..]) for the variable you want to include
sns.pairplot(tips)
```

```python
# sns.pairplot(tips,hue = 'sex') # color the dataset based on the column sex.
sns.pairplot(tips,hue = 'sex', palette = 'coolwarm')
```

```python
# rugplot() draws a dash mark for every single data point along the distribution lines. take one num arg
sns.rugplot(tips['total_bill'])
```

## 2 - Categorical plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
import numpy as np
```

```python
# the estimator takes a function that gives the actual aggregate grouping. By default is the mean. you can put your own funct
sns.barplot(x = 'sex', y = 'total_bill' , data = tips, estimator = np.std)
```

```python
# countplot
# the estimator counts the occurrence of the values. because of that we set only the x
sns.countplot(x='sex',data = tips)
```

```python
# boxplot the relationship of a numeric variable by a categorical variable
sns.boxplot(x = 'day',y = 'total_bill', data = tips) # shows the total bill per day
```

```python
# You can also add another layer of division, by using the hue arg
# Below we can see the total bill is split per day and and also by smokers
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
# violinplot shows the distribution of the data across some categories, same descrip as boxplot.
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True) # nice. try to remove tips and hue to see
# boxplot is recommended for non scientific audience
```

```python
# stripplot plots a scatter plot of a categorical variable and numeric variable. jitter arg is not required but 
# useful to have a better view
sns.stripplot(x='day',y = 'total_bill', data = tips, jitter = True, hue = 'sex', split = True)
```

```python
# swarplot combination of a stripplot and a violin plot. Not advised to use them
sns.swarmplot(x='day',y = 'total_bill', data = tips)
```

```python
# put violin and swarmplot on top of each other on the same figure. data exploration only not advised for presentation
sns.violinplot(x='day',y = 'total_bill', data = tips)
sns.swarmplot(x='day',y = 'total_bill', data = tips, color ='black')
```

```python
# factorplot the general plot. kind arg is what is going to describe the factorplot itself
sns.factorplot(x='day',y = 'total_bill', data = tips,kind = 'violin')
```

## 3 - Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
# heatmap are use to plot data that are in matrix form. annot arg for annototation, cmap arg for color
# matrix form: needs to have your data in a matrix fashion, pivot_table() does that.
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot = True, cmap = "coolwarm")
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
fp # great way to summarize dataset that have year and month
```

```python
sns.heatmap(fp) # as the years go by people are using planes a lot. and there are always a pick during the summer month
# june, july, august.
```

```python
sns.heatmap(fp, cmap = "magma")
```

```python
# linecolor: the color of the line, linewidth : the width of the line. they are use to have a better view.
# Read about these arg and others.
sns.heatmap(fp, cmap = "coolwarm", linecolor = 'white', linewidth = 4)
```

```python
#### cluster map: create hiercachical clustering of the heat map. cluster rows and columns based on their similarities
# cmap: stands for color mapping. standard_scale = 1 arg, normalize passengers.
sns.clustermap(fp, cmap = "coolwarm")
```

## 4 - Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# lmplot(): scatter plot with a regression line, hue arg for separation based on categorical feature
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
# can add matplotlib parameters like markers
# since hue has two levels, we need to give a list with two elts to the hue arg. 1elt for man and 2nd for female
# scatter_kws = {'s': a} to increase the size of the markers
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex',markers = ['o','v'],
          scatter_kws = {'s':100})
```

```python
# facetgrid for better separation
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col ='sex', row = 'time')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col ='day', hue = "sex")
```

```python
# we can change the size of the figure and aspect ratio
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col ='day', hue = "sex",
          aspect = 0.6, size = 8)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# g = FacetGrid(data,row = "nameofcatcol",col = "nameofcatcol") divides your grid by these two categories
# g.map(plotyouwant,numeric col you want to be mapped). you might need more arg depending of the type of plot you want to
# display, eg: scatter, boxplot needs two arg. you add them comma after comma.
g = sns.FacetGrid(data = tips, row = 'smoker', col = "time")
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data = tips, row = 'smoker', col = "time")
g.map(plt.scatter,'total_bill','tip')
```

## 6 - Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style(style = 'whitegrid') # change the background of your chart, by default style is set to darkgrid
sns.countplot(x = 'sex', data = tips)
```

#### <li>Set_style

```python
# You can also remove spines
sns.set_style(style = 'ticks') 
sns.countplot(x = 'sex', data = tips)
sns.despine() # by default it remove the top and right spines. if you want to remove the left and bottom change their values
# to true rather false, the default value.
```

```python
# You can also remove spines
sns.set_style(style = 'ticks') 
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True, bottom = True)
```

```python
plt.figure(figsize = (12,2)) # to change the size of your figure
sns.countplot(x = 'sex', data = tips)
```

#### <li>Set_context

```python
# You can also create a plot based on the context
sns.set_context(context = "poster", font_scale = 3)
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_context(context = "notebook") # the default context
sns.countplot(x = 'sex', data = tips)
```

#### <li>Palette and Color

```python
# notice that for palette to be applied you need to have the hue arg
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = "sex", palette = 'coolwarm')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = "sex", palette = 'seismic')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = "sex", palette = 'PuOr')
```
