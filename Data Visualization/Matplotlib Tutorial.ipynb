```python
import numpy as np
# linspace gives eleven evenly spaced numbers starting at 0 and ending at 5.
x = np.linspace(0,5,11)
y = x**2
```

### FUNCTIONAL

```python
plt.plot(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Title")
```

#### Multiples plot

```python
# first arg refers to the num of rows, 2nd number of columns, and 3rd the plot you are reffering to
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### OOP

```python
# 1st create a figure
fig = plt.figure()

# 2nd add a set of axes. it must be in the same cell
# pass a list: four argument(percent of that blank figre) left, bottom,width, height of the axes 
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#3rd: create the Plot. Remember, you use the axes to create the plot. you need to create the axes first from the figure as
# above
axes.plot(x,y)

# add labels
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
# Base axis. Start with it and then plot the other axes. Once you are done you can remove it.
# You always move from the base axis. left, bottom, width, height. (left,bottom) are the coordinate of the bootom left point of 
# a tgle. now to draw the tgle you need the width and the height. so from that point, you count the width and the height. Using
# the base axe as reference. DO NOT use the axes of any other plot.
base_axis = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.1,0.1,0.8,.8])
axes2 = fig.add_axes([0.6,0.2,0.3,0.3])
```

```python
# Draw figures 
# Create a figure
fig1 = plt.figure()

# Add two axes to the figures
axes1 = fig1.add_axes([0.1,0.1,0.8,.8])
axes2 = fig1.add_axes([0.2,0.5,.4,.3])

# Add plot on both axes
axes1.plot(x,y,'r')
axes1.set_title("LARGER PLOT")

axes2.plot(y,x,'b')
axes2.set_title("SMALLER PLOT")
```

# Part 2

### Subplots

```python
# The subplot creates automatically the figure and also the axes like we did above. The only difference it creates the axes
# based on the number of rows and columns. Notice that below we are doing some tuple(plt.subplots()is a tuple) unpacking
fig, axes = plt.subplots(nrows=1,ncols=2)
# axes in an array of axes, we have two axes, so we can iterate over too.
axes[0].plot(x,y)
axes[0].set_title("FIRST PLOT")

axes[1].plot(y,x)
axes[1].set_title("SECOND PLOT")

plt.tight_layout() # recommend to have at the end, to prevent overlapping
```

### Figure Size and DPI (aspect ratio)

```python
# figsize is one argument of figure, figsize takes a tuple(a,b) where a and b are the width and height of the figure in inches
# another argument of figure is dpi:dot per inches or pixels per inches, but we recommend to leave the default 
fig = plt.figure(figsize = (3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols = 1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# takes the name of the figure.the extension (png,jpg), you can also specify the dpi as another argument
fig.savefig('myfigure.png')
```

### Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# 
ax.plot(x,x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")

# when you call legend(), it looks in the indiv plot to see if there are some labels it can use.
# If it find them it use them. It does not find a label a warning apppears, saying no labelled objects found.
ax.legend() # it works, but in case the location of the legend overlaps the plot, use the loc argument which stands for location
# ax.legend(loc="best"= 0 you can write either one of them). ax.legend(loc = 0). Matplotlib will find the best location to 
# put the legend in the figure. you can also do ax.legend(loc = (0.1,0.1), which rep left, bottom) in case you did not like 
# any of the loc option matplotlib offers.
```

### Plot Appearance

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# add color
#axes.plot(x,y, color = "orange")

# Linewidth (lw works too rather than linewidth. default = 1) and linestyle (ls works too, 
# line visibility/transparency of the line (alpha)

# axes.plot(x,y,color = 'purple', linewidth = 3, linestyle = ":")

# Markers
axes.plot(x,y,color = 'purple', lw = 1, ls = "-",marker = "o",
          markersize = 20, markerfacecolor = "yellow", markeredgewidth = 3, markeredgecolor = "green")
```

### Control over axis appearance and Plot Range

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,color = 'purple', lw = 2, ls = "--")

# Set x and y limits(take a list of lower and upper bounds)
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
