Pandas plotting is for quick analysis, if you really want to customize your plots.

You should use matplotlib and seaborn(good for statistical plots).

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("df1",index_col = 0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv("df2")
```

```python
df2.head()
```

#### Histogram off of Pandas

```python
df1['A'].hist(bins=50, figsize = (8,6))
```

#### Area Plot

```python
df2.plot.area(alpha = 0.4)
```

#### Bar Plots

we should use categorical index with barplots

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

```python
df1['A'].plot.hist()
```

```python
df1.head()
```

#### Line Plot

```python
# good for stock. you can add as many matplotlib command you want
df1.plot.line(x=df1.index, y = "B",figsize = (12,3), lw = 1)
```

```python
# color(c) of dots is based on C
df1.plot.scatter(x="A", y="B",c='C')
```

```python
df1.plot.scatter(x="A", y="B",c='C',cmap = "coolwarm",figsize = (8,6))
```

```python
# we can also make the third dimensional arg be size(s). however it takes a dataframe column, not the string column
# and useful to multiply by a huge number if you column contain small decimal number
df1.plot.scatter(x="A", y="B",s = df1["C"]*100 ,figsize = (8,6))
```

```python
df2.plot.box()
```

```python
# create a 1000 by 2 data frame of normal random numbers
df = pd.DataFrame(np.random.randn(1000,2),columns = ['a','b'])
```

```python
df.head()
```

```python
# gridsize arg enlarges the size of the hexagons. the red hexagons imply that there are more points there.
df.plot.hexbin(x='a', y='b', gridsize = 25,figsize = (8,6),cmap = "coolwarm")
```

#### Kernel density plot

```python
# density() does the same thing as kde.
df['a'].plot.kde()
```

```python
# you can do off of the entire dataframe. In case you want to compare the density of the numeric columns at once
df2.plot.density()
```
