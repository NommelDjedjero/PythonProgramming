### 3. Create a lambda function that capitalizes strings

```python
def capString(x):
    if(type(x)==str):
        return x.upper()
    else:
        return x
#df1.loc[:,:].applymap(lambda x: capString(x)) 

# apply works on a row / column basis of a DataFrame, applymap works element-wise on a DataFrame, 
# and map works element-wise on a Series.
```

### 6. Did you notice the original dataframe is still lowercase? Why is that? Fix it and captalize Mjob and Fjob

```python
# because apply functions only act on a copy of the dataset not the real dataset.
df1[['Mjob','Fjob']] = df1[['Mjob','Fjob']].applymap(lambda x: capString(x))
```

### 7. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(x):
    if(x>20):
        return True
    else:
        return False
df1['legal_drinker'] = df1.age.apply(lambda x: majority(x))
```
