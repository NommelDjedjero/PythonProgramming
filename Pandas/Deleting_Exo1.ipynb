### Deleting: iris

```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
```

```python
iris.head()
```

### 1. Create columns for the dataset

```python
# create column names
column_names = ['sepal_length','sepal_width','petal_length','petal_width','class']
# set column names to the columns
iris.columns = column_names
```

```python
iris.head()
```

### 2. Is there any missing values in the dataset

```python
# iris.info() can also tells you which columns have a missing values as well as an overview of your dataset
iris.isnull().sum()
```

### 3. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
# We use the nan attribute from the numpy library
iris.loc[10:29,"petal_length"] = np.nan
iris.isnull().sum()
```

```python
iris.loc[10:29]  # or iris[10:30]
```

### 4. Good, now lets substitute the NaN values to 1.0

```python
iris.fillna(value = 1.0, inplace = True)
```

```python
iris.loc[10:29] # the missing petal_length is now fill with 1.0
```

### 5. Now let's delete the column class

```python
# must take at label and the axis. does not affect the dataframe 
iris.drop("class",axis = 1)
iris.head()
```

```python
# affect the dataframe with the inplace parameter
iris.drop("class",axis=1, inplace = True)
iris.head()
```

### 6. Set the first 3 rows to NaN

```python
iris.loc[0:2] = np.nan
```

```python
iris.head()
```

### 7. Delete the rows that have nan

```python
# axis = 0 tells that we are about to delete some rows, how = "any" means that we want to delete a row if it contains at least
# one nan in one of its cells. inplace = True, means we want to apply this change in our data frame. by def dropna has inplace = False
iris.dropna(axis = 0, how = "any",inplace = True)
```

```python
iris.head()
```

### 8. Reset the index so that it begins with 0 again

```python
# reset_index() creates a new index from 0, but also show a column index with the unresetted index.
# with drop = True, pandas drop the unressetted index.
iris = iris.reset_index(drop = True)
iris.head()
```
