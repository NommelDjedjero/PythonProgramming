### Filtering : Chipotle

```python
import numpy as np
import pandas as pd
```

```python
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

#### What is the price of each item

print a data frame with only two columns item_name and item_price

#### Sort by the name of the item

```python
#chipo.sort_values(by="item_name") # looks nice and returns a dataframe
chipo.item_name.sort_values() # Most used approach returns a series as you can check below
print(type(chipo.sort_values(by="item_name")))
print(type(chipo.item_name.sort_values()))
```

#### What was the quantity of the most expensive item ordered?

```python
# ANOTHER WAY OF TRANSFORMING PRICE INTO FLOAT
# clean the item_price column and transform it in a float
chipo.item_price = [float(value[1 : -1]) for value in chipo.item_price]
print(chipo.dtypes)
```

```python
# MY WAY OF TRANSFORMING PRICE INTO FLOAT 
print(chipo.dtypes)
type(chipo.item_price.str.replace("$","")) # series
chipo['item_price'] = chipo.item_price.str.replace("$","").astype(float)
print(chipo.dtypes)
```

```python
chipo.loc[chipo.item_price == chipo.item_price.max(),["item_name","quantity","item_price"]]
```

```python
chipo.loc[chipo.item_price == chipo.item_price.max(),"quantity"]
```

####  How many times were a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo.item_name == "Veggie Salad Bowl"].head() # count() is a series and df method
```

```python
chipo.loc[chipo.item_name == "Veggie Salad Bowl","item_name"].count()
```

#### How many times people orderd more than one Canned Soda

```python
chipo.loc[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)].head()
```

```python
chipo.loc[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1) , "item_name"].count()
```
