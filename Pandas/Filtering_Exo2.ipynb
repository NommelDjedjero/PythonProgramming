### 7. Filter teams that scored more than 6 goals

```python
euro12.loc[euro12.Goals > 6,["Team","Goals"]]
```

### 8. Select the teams that start with G

```python
euro12.loc[euro12.Team.str.startswith("G"),"Team"] # startswith() NOT startwith()
```

### 11. Present only the Shooting Accuracy from England, Italy and Russia

```python
# LONG WAY
euro12.loc[(euro12.Team == "England")|(euro12.Team == "Italy")|(euro12.Team == "Russia"),["Team","Shooting Accuracy" ]]
```

```python
# SHORT WAY
euro12.loc[euro12.Team.isin(["England","Italy","Russia"]),["Team","Shooting Accuracy"]]
```
